```python
def alt_harmonic_series(n):
    sum = 0
    for i in range(1, n + 1):
        sum += (-1) ** (i + 1) * (1 / i)
    print(sum)
```

```python
alt_harmonic_series(10000)
```

```python
x = np.linspace(1, 100, 10000)
y = np.sqrt(2 * x - 1) / (x + 12)

plt.plot(x, y)
plt.show()
```
